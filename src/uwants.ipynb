```python
import requests
from bs4 import BeautifulSoup, NavigableString, Tag
import re
import os.path
import pandas as pd
```

```python
# input the number of pages you want to scrape
# the data will be automatically saved as a csv file
def collectThreadURL(page_num):
    src_url = "https://news.discuss.com.hk/forumdisplay.php?fid=54&threadMode=neutral&from=2&page="

    responseNums = []
    threadLinks = []

    for i in range(1, page_num+1):
        r = requests.get(src_url+str(i))
        soup = BeautifulSoup(r.text, 'html.parser')

        reactionNums_raw = soup.find_all(class_="nums")
        num = 0
        for n in reactionNums_raw:
            if num > 4 and num != 8:
                txt_raw = n.text
                txt = txt_raw.split(" ")[0][1:]
                responseNums.append(txt)
            num+=1

        tsubject = soup.find_all(class_="tsubject")
        threadLinks_per_page = []
        for link in tsubject:
            thread = link.find('a')
            thread_link = thread.get("href")
            threadLinks_per_page.append(thread_link)
        threadLinks.extend(threadLinks_per_page[3::])

    col = ["response_num", "thread_link"]
    data = pd.DataFrame(list(zip(responseNums, threadLinks)), columns = col)
    data_file_name = "uwants_links.csv"
    data.to_csv(data_file_name,index=False)
    return data_file_name
```

```python
def collectPageDiscussion(soup, n):
    spans = soup.find_all("span", {"id": re.compile(r'^postorig')})
    discussions = ""
    for span in spans:
        discussions += "\n===============================\n"+span.text
        n+=1
    return discussions, n
```

```python
def collectNextPageURL(soup):
    pages = soup.find(class_="pagination-buttons")
    next_page = pages.find("a").get("href")
    return next_page
```

```python
def collectThreadDiscussion(threadURL, setMax=False, maxPageNum=-1):
    thread_url_prefix = "https://news.discuss.com.hk/"
    r = requests.get(thread_url_prefix+threadURL)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    n = 1
    pages = soup.find(class_="pagination-buttons")
    total_page_num = int(pages.find("em").text[1:])
    next_page = collectNextPageURL(soup)
    
    if setMax:
        total_page_num = maxPageNum
    
    discussions, n = collectPageDiscussion(soup, n)
    while total_page_num > 1:
        r = requests.get(thread_url_prefix+next_page)
        soup = BeautifulSoup(r.text, 'html.parser')
        discussion_new = ""
        n_new = 0
        discussion_new, n_new == collectPageDiscussion(soup, n)
        discussions += discussion_new
        n += n_new 
        next_page = collectNextPageURL(soup)
        print(total_page_num)
        total_page_num-=1
    
    return discussions
```

```python
def utilizeCollectedURLs(data_file_name):
    data = pd.read_csv("..links/"+data_file_name)
    
    response_num_sorted = data["response_num"].sort_values().tolist()
    index = int(len(response_num_sorted)*0.8)
    threshold = response_num_sorted[index]
    
    n = 1
    for i, row in data.iterrows():
        response_num = int(row["response_num"])
        if response_num > threshold:
            discussions = collectThreadDiscussion(row["thread_link"])
            save_path = os.getcwd()+"../threads/"
            discussions_file_name = "thread"+str(n)+".txt"
            fileName = os.path.join(save_path, discussions_file_name)
            with open(fileName, "w", encoding="utf-8") as text_file:
                text_file.write(discussions)
            print(n, "finished")
            n+=1
```

```python
def mine_data(page_num):
    
    data_file_name = collectThreadURL(page_num)
    
    utilizeCollectedURLs(data_file_name)
```
